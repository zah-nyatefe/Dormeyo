```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

dataset = pd.read_csv("red_wine.csv")
```

```python
data= dataset['alcohol']
plt.figure(figsize=(10,8))
plt.hist(data, bins=15,color='blue',edgecolor='black')
plt.title('Histogram of Alcohol Content in Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')

plt.show()
```

```python
quality_counts=dataset['quality'].value_counts()

plt.figure(figsize=(8,6))
plt.pie(quality_counts, labels=quality_counts.index,autopct='%1.1f%%',colors=plt.cm.Paired.colors,startangle=90)

plt.title('Wine Quality Qistribution')
plt.show()
```

```python
x= dataset['quality']
y= dataset['residual sugar']
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.title('Gragh of quality vs residual sugar')
plt.xlabel('Quality')
plt.ylabel('Residual Sugar')
plt.show()
```

```python
alcohol_content= dataset['alcohol']
ph= dataset['pH']
plt.scatter(alcohol_content, ph,color='blue', alpha=0.5, edgecolor='black')

plt.title('Alcohol Content vs.pH of Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('pH')
plt.show()
```

 Report on the plots and graphs obtained
 ---
 ### Histogram of Alcohol Content:
 - The histogram shows the distribution of alcohol content in the dataset.It indicates that most wines have alcohol content between a few outliers having or lower alcohol content.
 ### Scatter Plot of Alcohol vs. Quality:
 - The scatter plot illustrates the relationship between alcohol content and quality. It suggests a potential positives correlation, where wines with higher alcohol content tend to have higher quality ratings.
